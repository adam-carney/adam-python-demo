```python
# BEGIN: Modify first_name column based on the given condition
master_test_df['first_name'] = master_test_df['first_name'].apply(lambda x: x[0] if x[0] in ['A', 'D', 'F', 'R'] else x)
# END: Modify first_name column based on the given condition

master_test_df.head()
```

```python
# BEGIN: Reverse month and day fields for specific birthdates
def reverse_month_day(date):
    if date.month in [12, 6, 2] and date.day < 12:
        return date.replace(month=date.day, day=date.month)
    return date

master_test_df['birthdate'] = master_test_df['birthdate'].apply(reverse_month_day)
# END: Reverse month and day fields for specific birthdates

master_test_df.head()
```

```python
import re

def reverse_numbers(address):
    numbers = re.findall(r'\b\d+\b', address)
    for number in numbers:
        if 0 <= int(number) <= 200:
            reversed_number = number[::-1]
            address = address.replace(number, reversed_number)
    return address

master_test_df['street_address'] = master_test_df['street_address'].apply(reverse_numbers)

master_test_df.head()
```

```python
def adjust_street_address(address):
    numbers = re.findall(r'\b\d+\b', address)
    for number in numbers:
        if 6000 <= int(number) <= 7000:
            adjusted_number = str(int(number) - 30)
            address = address.replace(number, adjusted_number)
    return address

master_test_df['street_address'] = master_test_df['street_address'].apply(adjust_street_address)

master_test_df.head()
```
